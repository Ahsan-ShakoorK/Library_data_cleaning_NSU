```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
data11 = pd.read_csv('data/data.csv')
print(data11.shape)
pd.set_option('display.max_columns', None)
data11.head(1)
```

```python
df = data11
def clean_years(year_string):
    if pd.isna(year_string):
        return None, None, None  # Return None for all if year is missing
    
    # Remove corrupted characters (non-ASCII, like "пїЅ", etc.)
    year_string = re.sub(r'[^\x00-\x7F]+', '', str(year_string))  # Ensure it's a string and remove non-ASCII characters
    
    # Extract all years (four digits after ^D and any parenthetical years)
    years = re.findall(r'\^D(\d{4})|(\d{4})', year_string)  
    
    # Flatten the list and remove empty strings
    years = [year for sublist in years for year in sublist if year]
    
    if len(years) > 1:
        return None, years, "Multiple years found"  # Return multiple years and issue message
    elif len(years) == 1 and len(years[0]) == 4:
        return years[0], None, None  # Return single valid year with no issues
    else:
        return None, None, "Invalid or incomplete year"  # No valid year found

# Apply the updated function and create new columns
df[['cleaned_year_1', 'multiple_years', 'year_issues']] = df['210'].apply(lambda x: pd.Series(clean_years(x)))

# Display the first 15 rows of the updated DataFrame
df[['210', 'cleaned_year_1', 'multiple_years', 'year_issues']].sample(2)
```
