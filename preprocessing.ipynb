```python
import pandas as pd
import re


df = data11

def clean_years(year_string):
    if pd.isna(year_string):
        return None, None  # Return None for both columns if year is missing
    
    # Remove corrupted characters (non-ASCII, like "пїЅ", etc.)
    year_string = re.sub(r'[^\x00-\x7F]+', '', str(year_string))  # Ensure it's a string and remove non-ASCII characters
    
    # Extract all years (four digits after ^D and any parenthetical years)
    years = re.findall(r'\^D(\d{4})|(\d{4})', year_string)  
    
    years = [year for sublist in years for year in sublist if year]
    
    if len(years) > 1:
        return years, "Multiple years found"
    elif len(years) == 1 and len(years[0]) == 4:
        return years[0], None 
    else:
        return None, "Invalid or incomplete year"

df[['cleaned_year', 'year_issues']] = df['210'].apply(lambda x: pd.Series(clean_years(x)))

print(df[['210', 'cleaned_year', 'year_issues']].head())  # Display the original and cleaned year columns
print(f"Number of Nulls in 'cleaned_year': {df['cleaned_year'].isna().sum()}")  # Check for missing data in cleaned_year
print(f"Number of Issues in 'year_issues': {df['year_issues'].notna().sum()}")  # Count of issues
```

```python
print(f"Number of Nulls in '210': {df['210'].isna().sum()}") 
print(f"Number of Nulls in 'cleaned_year': {df['cleaned_year'].isna().sum()}") 
print(f"Number of Issues in 'year_issues': {df['year_issues'].notna().sum()}") 
```
